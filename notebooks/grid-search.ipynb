# Turing the hypter-parameters of RNN with Grid Search

```python
from __future__ import absolute_import
from __future__ import division
from __future__ import print_function
```

```python
import sys
sys.path.append('../src/')
```

```python
import numpy as np
seed = 1234
np.random.seed(seed)
import pandas as pd
```

```python
from sklearn.utils import shuffle
from sklearn.model_selection import GridSearchCV
```

```python
from keras.wrappers.scikit_learn import KerasClassifier
```

```python
from gestureai import load_dataset, rnn
```

```python
input_dim = 3
timesteps = 40
n_classes = 5
n_files_class = 200
batch_size = 20
epochs = 50
n_cv = 4
```

# Load Datasets

```python
X = load_dataset('../datasets/gesture-3axis-accel.tar.gz')
```

```python
y = np.array([])
for i in range(n_classes):
    y = np.r_[y, [i] * n_files_class]
```

```python
X, y = shuffle(X, y)
```

# Execute Grid Search

```python
param_grid = {
    "rnn_cell": ['LSTM', 'GRU'],
    "n_hidden": [128, 256, 512, 1024,],
}
```

```python
model = KerasClassifier(build_fn=rnn, nb_epoch=epochs, batch_size=batch_size, verbose=1)
```

```python
clf = GridSearchCV(estimator=model, param_grid=param_grid, cv=n_cv, scoring='accuracy')
```

```python
res = clf.fit(X, y)
```

# Results

```python
res.cv_results_
```

```python
"Hyper Parameters:", res.best_params_
```

```python
"Accuracy:", res.best_score_
```
